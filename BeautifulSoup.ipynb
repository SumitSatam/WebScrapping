```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_banks_in_India'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

## Extracting 1st Table

```python
table = soup.find('table', class_ = 'wikitable sortable')
```

```python
tab_titles = table.find_all('th')
```

```python
tab_titles
```

```python
all_table_titles = [title.text.strip() for title in tab_titles]

print(all_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = all_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]

  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```
